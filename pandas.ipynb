```python
# Groupby mainly refers to a process involving one or more of the following steps:
# Splitting: Split data into groups, by applying some conditions on datasets.
# Applying: Apply a function to each group independently.
# Combining: Combine different datasets after applying groupby
```

```python
data1 = {
    'Name': ['Jai', 'Anuj', 'Jai', 'Princi', 'Gaurav', 'Anuj', 'Princi', 'Abhi'], 
    'Age': [27, 24, 22,32, 33, 36, 27, 32], 
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj', 'Jaunpur', 'Kanpur', 
                'Allahabad', 'Aligarh'], 
    'Qualification': ['Msc', 'MA', 'MCA', 'Phd', 'B.Tech', 'B.com', 'Msc', 'MA']
}
df = pd.DataFrame(data1)
print("Original DataFrame")
print(df)
gk = df.groupby('Name')
print("After creating Groups")
gk.first()
```

```python
# Aggregation is a process in which we compute a summary statistic (or statistics) about each group. 
# e.g.: Compute group sums or means
result = gk['Age'].aggregate('sum')
print(result)
```

```python
# Concatenation
data1 = {
    'key': ['K0', 'K1', 'K2', 'K3'],
    'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
    'Age': [27, 24, 22, 32]
}
data2 = {
    'key': ['K0', 'K1', 'K2', 'K3'],
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
    'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']
}
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)
display(df, df1)
res  = pd.concat([df, df1], axis=1)
print(res)
# axis=1: Concatenates along the columns (adding columns side by side, horizontally).
```

```python
# Merge
res  = pd.merge(df, df1, on='key')
print(res)
```

```python
# Join, used for combining the columns of two potentially differently 
# indexed DataFrames into a single result DataFrame.
# Syntax: separator.join(iterable)
# separator: The string placed between elements of the iterable.
# iterable: A sequence of strings (e.g., list, tuple, etc..) to join together.

data1 = {
    'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
    'Age': [27, 24, 22, 32]
}
data2 = {
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
    'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']
}
df = pd.DataFrame(data1, index=['K0', 'K1', 'K2', 'K3'])
df1 = pd.DataFrame(data2, index=['K0', 'K2', 'K3', 'K4'])
display(df, df1)
res  = df.join(df1)
print(res)
```
