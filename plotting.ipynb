```python
# Exercise 1: line Chart with Matplotlib
# Plot a simple line chart using Matplotlib showing 
# the population growth from 2000 to 2020.
# Note that the population growth formula is: 2.5+(1/10)*x, 
# where x represents the year number: 1st, 2nd, ... 20.
# Hint: you need to create 2 lists: one for the range of 
# years and another one for the population values according 
# to the given formula, then plot the line chart.

import matplotlib.pyplot as plt

years = [_ for _ in range(2000, 2021)]
print(years)
population = [2.5+(1/10)*x for x in range(1, len(years)+1)]
print(population)
plt.ylabel('Population')
plt.xlabel('Year')
plt.title('Population Growth')
plt.plot(years, population, 'green', marker='o', label='growth')
plt.legend(loc="lower right")
plt.show()
```

```python
# Load the IRIS.csv dataset from Seaborn and display its first five rows
# Create a scatter plot of SepalLengthCm vs. SepalWidthCm and explain the graph
# Create a scatter plot of PetalLengthCm vs. PetalWidthCm and explain the graph
# Plot the histogram for the SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm 
# columns and explain each graph
# Create a box plot to visualize the distribution of SepalLengthCm grouped by species 
# and explain each graph
# Create a box plot to visualize the distribution of PetalLengthCm grouped by species 
# and explain each graph
# Generate a pair plot for the Iris dataset that compares each pair of variants according 
# to a third one: species and explain in details each sub plot
# Create a heatmap of the correlation matrix for the Iris dataset and explain the plotting result
# Display a regression plot for SepalLengthCm vs. PetalLengthCm and explain the plotting result
# Display a regression plot for SepalWidthCm vs. PetalWidthCm and explain the plotting result
# Create a count plot to show the number of observations for each species in the Iris datset. Explain
# Create a Kernel Densit Estimate (KDE) plot for the distribution of SepalLengthCm. Explain
# Use seaborn to plot all the attributes of the dataset by normalizing the species attributes. Explain

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the Iris dataset
df = sns.load_dataset('iris')

print("Original Iris dataset from seaborn:")
print(df.head())

column_mapping = {
    'sepal_length': 'SepalLengthCm',
    'sepal_width': 'SepalWidthCm',
    'petal_length': 'PetalLengthCm',
    'petal_width': 'PetalWidthCm',
    'species': 'Species'
}
df = df.rename(columns=column_mapping)

# 1. Scatter plot: SepalLengthCm vs SepalWidthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.grid(True)
plt.show()
print("Strong correlation between x and y since there are two clusters and not 50,\
    and existence of outliers on the extremes")

# 2. Scatter plot: PetalLengthCm vs PetalWidthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data=df)
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.grid(True)
plt.show()
print("Strong correlation between x and y since there are two clusters and not 50")

# 3. Histograms for all measurement columns
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.histplot(df['SepalLengthCm'], kde=True, color='blue')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.histplot(df['SepalWidthCm'], kde=True, color='green')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
sns.histplot(df['PetalLengthCm'], kde=True, color='red')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
sns.histplot(df['PetalWidthCm'], kde=True, color='purple')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# 4. Box plot for SepalLengthCm by Species
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=df, legend=False)
plt.title('Box Plot: Sepal Length by Species')
plt.grid(True)
plt.show()

# 5. Box plot for PetalLengthCm by Species
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='PetalLengthCm', data=df, legend=False)
plt.title('Box Plot: Petal Length by Species')
plt.grid(True)
plt.show()

# 6. Pair plot
plt.figure(figsize=(12, 10))
pair_plot = sns.pairplot(df, hue='Species', palette='viridis')
plt.suptitle('Pair Plot of Iris Dataset Features by Species', y=1.02, fontsize=16)
plt.show()

# 7. Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.drop('Species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap for Iris Dataset')
plt.show()

# 8. Regression plot: SepalLengthCm vs PetalLengthCm
plt.figure(figsize=(10, 6))
sns.regplot(x='PetalLengthCm', y='SepalLengthCm', data=df, scatter_kws={'alpha':0.5})
plt.title('Regression Plot: Sepal Length vs Petal Length')
plt.grid(True)
plt.show()

# 9. Regression plot: SepalWidthCm vs PetalWidthCm
plt.figure(figsize=(10, 6))
sns.regplot(x='PetalWidthCm', y='SepalWidthCm', data=df, scatter_kws={'alpha':0.5})
plt.title('Regression Plot: Sepal Width vs Petal Width')
plt.grid(True)
plt.show()

# 10. Count plot for Species
plt.figure(figsize=(10, 6))
sns.countplot(x='Species', data=df, hue='Species', palette='viridis', legend=False)
plt.title('Count of Observations by Species')
plt.grid(axis='y')
plt.show()

# 11. KDE plot for SepalLengthCm
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='SepalLengthCm', hue='Species', fill=True, common_norm=False, palette='viridis', alpha=0.7)
plt.title('KDE Plot: Distribution of Sepal Length by Species')
plt.grid(True)
plt.show()

# 12. Normalized radar chart for species attributes
numeric_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Calculate the mean for each feature by species
means_by_species = df.groupby('Species')[numeric_cols].mean()

min_vals = means_by_species.min()
max_vals = means_by_species.max() 
normalized_means = (means_by_species - min_vals) / (max_vals - min_vals)

# Create radar chart
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, polar=True)

# Set the angles for each feature (in radians)
angles = np.linspace(0, 2*np.pi, len(numeric_cols), endpoint=False).tolist()
angles += angles[:1]  # Close the loop

# Add the feature names as labels
plt.xticks(angles[:-1], numeric_cols, fontsize=12)

# Plot each species
for i, species in enumerate(normalized_means.index):
    values = normalized_means.loc[species].values.tolist()
    values += values[:1]  # Close the loop
    ax.plot(angles, values, linewidth=2, label=species)
    ax.fill(angles, values, alpha=0.25)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title('Normalized Features by Species', size=15)
plt.show()
```
