```python
# Exercise 1: line Chart with Matplotlib
# Plot a simple line chart using Matplotlib showing 
# the population growth from 2000 to 2020.
# Note that the population growth formula is: 2.5+(1/10)*x, 
# where x represents the year number: 1st, 2nd, ... 20.
# Hint: you need to create 2 lists: one for the range of 
# years and another one for the population values according 
# to the given formula, then plot the line chart.

import matplotlib.pyplot as plt

years = list(range(2000, 2021))
print(years)
population = [2.5 + (1 / 10) * x for x in range(1, len(years) + 1)]
print(population)
plt.ylabel('Population')
plt.xlabel('Year')
plt.title('Population Growth')
plt.plot(years, population, 'green', marker='o', label='growth')
plt.legend(loc="lower right")
plt.show()
```

```python
# Exercise 5: Create the following list:
# Departments = ["HR", "IT", "Finance", "Marketing", "Sales"] * 5.
# Create Salary list which contains 25 random numbers between 40,000 and 100,000.
# Create a dataframe using these 2 lists, print the dataframe and show a box plot 
# comparing salaries in different departments.

import numpy as np
import pandas as pd
np.random.seed(0)
departments = ["HR", "IT", "Finance", "Marketing", "Sales"] * 5
salaries = np.random.randint(40_000, 100_001, 25)
df = pd.DataFrame({
    'Department': departments,
    'Salary': salaries
})
print(df)
res = df.groupby('Department')['Salary'].agg('sum')
print(res)
df.boxplot(column='Salary', by='Department')
plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Salary Distribution by Department")
plt.show()
```

```python
# Exercise 6:
# Create a square matrix M of dimension 10*10 M(Xij, where X is a 
# random value between 16 and 36).
# Print M and check if all values are in the right range.
# Visualize correlations in a dataset using a heatmap, note that the 
# cmap should utilize the coolwarm value
# Discuss the heatmeap result.

import numpy as np

np.random.seed(0)
M = np.random.randint(16, 37, (10, 10))
print(M)
in_range = lambda arr: all(16 <= val < 37 for val in [val for x in arr for val in x]) # arr.flatten()
result = "Values are in the right range" if in_range(M) else "Values are not in the right range"
print(result)
plt.imshow(M, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Heatmap")
plt.show()
```

```python
# Exercise 7:
# Plot a simple line chart using Matplotlib showing the population growth 
# from 1980 to 2020. Note that the population growth formula is 
# ((2.1 + (1/10) * x^2) / (x^2 + 1)), where x represent the year number: 1st, 2nd, , 40th.
# Hint: you need to create 2 lists: one for the range of years and another one for the population
# values according to the given formula, Print the content of both lists then plot the line chart.

x_axis = list(range(1980, 2021))
print(x_axis)
y_axis = [((2.1 + 0.1 * x ** 2) / (x ** 2 + 1)) for x in range(1, len(x_axis)+1)]
print(y_axis)
plt.xlabel('Year')
plt.ylabel('Population')
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Exercise 9: Comparing Real and Simulated Weather Data
# Load a real CSV dataset of recorded temperatures and compare it with simulated temperature (created randomly) variations.

df = pd.read_csv('Datasets/weather.csv')
print(df)
df['Date'] = pd.to_datetime(df['Date'])
simulated_temps = df['Temperature'] + np.random.normal(0, 2, len(df))
# Simulating variations
plt.plot(df['Date'], df['Temperature'], label="Actual Data", color='blue')
plt.plot(df['Date'], simulated_temps, label="Simulated Data", color='red', linestyle='dashed')
plt.title("Weather Data Comparison")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Load the IRIS.csv dataset from Seaborn and display its first five rows
# Create a scatter plot of SepalLengthCm vs. SepalWidthCm and explain the graph
# Create a scatter plot of PetalLengthCm vs. PetalWidthCm and explain the graph
# Plot the histogram for the SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm 
# columns and explain each graph
# Create a box plot to visualize the distribution of SepalLengthCm grouped by species 
# and explain each graph
# Create a box plot to visualize the distribution of PetalLengthCm grouped by species 
# and explain each graph
# Generate a pair plot for the Iris dataset that compares each pair of variants according 
# to a third one: species and explain in details each sub plot
# Create a heatmap of the correlation matrix for the Iris dataset and explain the plotting result
# Display a regression plot for SepalLengthCm vs. PetalLengthCm and explain the plotting result
# Display a regression plot for SepalWidthCm vs. PetalWidthCm and explain the plotting result
# Create a count plot to show the number of observations for each species in the Iris datset. Explain
# Create a Kernel Densit Estimate (KDE) plot for the distribution of SepalLengthCm. Explain
# Use seaborn to plot all the attributes of the dataset by normalizing the species attributes. Explain

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
df = sns.load_dataset('iris')

print(df.head())

# 1. Scatter plot: SepalLengthCm vs SepalWidthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=df)
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()
print("Strong correlation between x and y since there are two clusters and not 50,\
    and existence of outliers on the extremes")

# 2. Scatter plot: PetalLengthCm vs PetalWidthCm
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal_length', y='petal_width', data=df)
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.show()
print("Strong correlation between x and y since there are two clusters and not 50")

# 3. Histograms for all measurement columns
df.hist(figsize=(10,8), bins=20)
plt.suptitle("Histogram of Sepal and Petal Dimensions")
plt.show()

# 4. Box plot for SepalLengthCm by Species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Box Plot: Sepal Length by Species')
plt.show()

# 5. Box plot for PetalLengthCm by Species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Box Plot: Petal Length by Species')
plt.show()

# 6. Pair plot
plt.figure(figsize=(12, 10))
pair_plot = sns.pairplot(df, hue='species')
plt.suptitle('Pair Plot of Iris Dataset Features by Species', y=1.02, fontsize=16)
plt.show()

# 7. Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap for Iris Dataset')
plt.show()

# 8. Regression plot: SepalLengthCm vs PetalLengthCm
plt.figure(figsize=(10, 6))
sns.regplot(x='petal_length', y='sepal_length', data=df)
plt.title('Regression Plot: Sepal Length vs Petal Length')
plt.show()

# 9. Regression plot: SepalWidthCm vs PetalWidthCm
plt.figure(figsize=(10, 6))
sns.regplot(x='petal_width', y='sepal_width', data=df, scatter_kws={'alpha':0.5})
plt.title('Regression Plot: Sepal Width vs Petal Width')
plt.show()

# 10. Count plot for Species
plt.figure(figsize=(10, 6))
sns.countplot(x='species', data=df)
plt.title('Count of Observations by Species')
plt.show()

# 11. KDE plot for SepalLengthCm
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='sepal_length', shade=True)
plt.title('KDE Plot: Distribution of Sepal Length by Species')
plt.show()
```
