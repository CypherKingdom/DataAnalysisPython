```python
# Seaborn Pairplot
# Syntax: seaborn.pairplot(data, **kwargs)
# data: Tidy(long-form) dataframe where each column is a variable 
# and each row is an observation
# hue: Variable in "data" to map plot aspects to differnt colors
# palette: dict or seaborn color palette
# {x, y}: lists of variable names, optional
# dropna: boolean, optional

import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
df.head()
sns.pairplot(df)
plt.show()
```

```python
# Plotting Selected Variables
selected_vars = ['total_bill', 'tip']
sns.pairplot(df, vars=selected_vars)
plt.show()
```

```python
# Using 'hue' parameter
sns.pairplot(df, vars=selected_vars, hue='size')
plt.show()
```

```python
# Using 'palette' parameter always used with 'hue'
sns.pairplot(df, vars=selected_vars, hue='size', palette='husl')
plt.show()
```

```python
# Using 'diag_king' parameter
sns.pairplot(df, diag_kind='kde')
plt.show()
# 'diag_kind' parameter specifites the type of plot to display along the diagonal axis,
# representing the univariant distribution of each variable
# options include 'hist' for histograms, 'kde' for kermel density estimates,
# and 'scatter' for scatter plots.
# Choose based ont he nature of the data and analysis goals.
```

```python
# Using 'kind' parameter
sns.pairplot(df, kind='reg')
plt.show()
```

```python
# Using 'markers' parameter allows to specify different markers for different categories
sns.pairplot(df, hue='sex', markers=["o", "s"])
plt.show()
# This function is great for exploring relationships between multiple numerical variables 
# in your dataset while visually encoding categorical data.
```

```python
# For advanced customization, you can access the underlying FaceGrid object and
# modify it further.
g = sns.pairplot(df, hue='day')
g.fig.suptitle("Pairplot of Tips Dataset", y=1.02) # Add title
g.set(xticks=[], yticks=[]) # Remove tick labels
plt.show()
```
